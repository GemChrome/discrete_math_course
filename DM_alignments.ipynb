```python
#1 Find the global alignment
first = input()
second = input()

def alignment(first, second):
    m = -1
    g = -2
    scoring = [[0] * (len(second) + 1) for _ in range(len(first) + 1)]
    path = [['-'] * (len(second) + 1) for _ in range(len(first) + 1)]

    for i in range(len(first) + 1):
        for j in range(len(second) + 1):
            scoring[0][0] = 0
            path[0][0] = '-'
            if i == 0 and j > 0:
                scoring[0][j] = g * j
                path[0][j] = 'left'
            elif j == 0 and i > 0:
                scoring[i][0] = g * i
                path[i][0] = 'up'
            else:
                match = scoring[i - 1][j - 1] + 1 if first[i - 1] == second[j - 1] else scoring[i - 1][j - 1] + m
                insertion = scoring[i][j - 1] + g
                deletion = scoring[i - 1][j] + g
                scoring[i][j] = max(match, insertion, deletion)

                if scoring[i][j] == match:
                    path[i][j] = 'DIAG'
                elif scoring[i][j] == deletion:
                    path[i][j] = 'up'
                else:
                    path[i][j] = 'left'

    i = len(first)
    j = len(second)
    x = []
    y = []
    while i > 0 or j > 0:
        if path[i][j] == 'DIAG':
            x.append(first[i - 1])
            y.append(second[j - 1])
            i -= 1
            j -= 1
        elif path[i][j] == 'up':
            x.append(first[i - 1])
            y.append('-')
            i -= 1
        elif path[i][j] == 'left':
            x.append('-')
            y.append(second[j - 1])
            j -= 1

    x.reverse()
    y.reverse()
    return x, y

res = alignment(first, second)
print(''.join(res[0]))
print(''.join(res[1]))
```

```python
#3 Find the global alignment with gap opening and extension penalty
first = input()
second = input()
openPenalty, extPenalty = map(int, input().split())

def affine_gap(first, second, openPenalty, extPenalty):
    m = -1

    maintable = [[0] * (len(first) + 1) for _ in range(len(second) + 1)]
    maintable[0][0] = 0
    maintable[0][1] = openPenalty + extPenalty
    maintable[1][0] = openPenalty + extPenalty
    for i in range(2, len(maintable)):
        for j in range(2, len(maintable[0])):
            maintable[i][0] = maintable[i - 1][0] + extPenalty
            maintable[0][j] = maintable[0][j - 1] + extPenalty

    upper = [[0] * (len(first) + 1) for _ in range(len(second) + 1)]
    upper[1][0] = openPenalty + extPenalty
    for j in range(len(upper[0])):
        for i in range(2, len(upper)):
            upper[0][j] = len(first) * extPenalty * 1000000
            upper[i][0] = upper[i - 1][0] + extPenalty

    lower = [[0] * (len(first) + 1) for _ in range(len(second) + 1)]
    lower[0][1] = openPenalty + extPenalty
    for i in range(len(lower)):
        for j in range(2, len(lower[0])):
            lower[i][0] = len(first) * openPenalty * 1000000
            lower[0][j] = lower[0][j - 1] + extPenalty

    path_main = [[0] * (len(first) + 1) for _ in range(len(second) + 1)]
    path_main[0][0] = '-'
    for i in range(len(path_main)):
        for j in range(len(path_main[0])):
            path_main[i][0] = 'U'
            path_main[0][j] = 'L'

    path_upper = [[0] * (len(first) + 1) for _ in range(len(second) + 1)]
    path_upper[0][0] = '-'
    for i in range(len(path_upper)):
        for j in range(len(path_upper[0])):
            path_upper[i][0] = 'U'
            path_upper[0][j] = '-'

    path_lower = [[0] * (len(first) + 1) for _ in range(len(second) + 1)]
    path_lower[0][0] = '-'
    for i in range(len(path_lower)):
        for j in range(len(path_lower[0])):
            path_lower[i][0] = '-'
            path_lower[0][j] = 'L'

    for i in range(1, len(second) + 1):
        for j in range(1, len(first) + 1):
            lower[i][j] = max(lower[i][j - 1] + extPenalty, maintable[i][j - 1] + openPenalty + extPenalty)
            if lower[i][j] == lower[i][j - 1] + extPenalty:
                path_lower[i][j] = 'L'
            else:
                path_lower[i][j] = 'M'

            upper[i][j] = max(upper[i - 1][j] + extPenalty, maintable[i - 1][j] + openPenalty + extPenalty)
            if upper[i][j] == upper[i - 1][j] + extPenalty:
                path_upper[i][j] = 'U'
            else:
                path_upper[i][j] = 'M'

            choice = 1 if first[j - 1] == second[i - 1] else m
            maintable[i][j] = max(maintable[i - 1][j - 1] + choice, lower[i][j], upper[i][j])
            if maintable[i][j] == maintable[i - 1][j - 1] + choice:
                path_main[i][j] = 'M'
            elif maintable[i][j] == lower[i][j]:
                path_main[i][j] = 'L'
            else:
                path_main[i][j] = 'U'

    x = []
    y = []
    i = len(second)
    j = len(first)
    while i > 0 or j > 0:
        if path_main[i][j] == 'M':
            x.append(first[j - 1])
            y.append(second[i - 1])
            i -= 1
            j -= 1
        elif path_main[i][j] == 'L':
            while path_lower[i][j] == 'L' and j > 0:
                x.append(first[j - 1])
                y.append('-')
                j -= 1
            if path_lower[i][j] == 'M':
                x.append(first[j - 1])
                y.append('-')
                j -= 1

        elif path_main[i][j] == 'U':
            while path_upper[i][j] == 'U' and i > 0:
                x.append('-')
                y.append(second[i - 1])
                i -= 1
            if path_upper[i][j] == 'M':
                x.append('-')
                y.append(second[i - 1])
                i -= 1

    x.reverse()
    y.reverse()
    return x, y

res = affine_gap(first, second, openPenalty, extPenalty)
print(''.join(res[0]))
print(''.join(res[1]))
```
